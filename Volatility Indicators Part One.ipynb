# Volatility Indicators Part I

Measuring "volatility" just means measuring how much something "moves" during a given time frame. 

A simple way to measure volatility is to take the average rate of change. 

```python
day = ['Monday','Tuesday','Wednesday','Thursday','Friday']
price = [102,104,102,106,104]
data = pd.DataFrame(price, index=day,columns=['price'])
data.plot(figsize=(8,6))
plt.show()
```

The rate of change is calculated

```python
data['change'] = data['price'] - data['price'].shift(1)
```

A common method to calculate volatility for futures traders is "Average True Range" (ATR). Originally developed by J. Welles Wilder, Jr. in his book "New Concepts in Technical Trading Systems" (1978).

The motivation for this calculation is to measure the degree of price volatility, it is not an indication of price trend.

The range of a day's trading is the high minus the low, and the "true range" extends it to yesterday's closing price if it was outside of today's range.

$$TR = max[(high - low), abs(high-close[1]), abs(low-close[1])]$$
